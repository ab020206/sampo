```python
# PS 18 - Heart Disease Visualization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read dataset
df = pd.read_csv("/content/Cleavland.csv", header=None)

# Add column names
df.columns = [
    "age","sex","cp","trestbps","chol","fbs",
    "restecg","thalach","exang","oldpeak",
    "slope","ca","thal","target"
]

# a. HISTOGRAM

plt.figure(figsize=(6,4))

plt.hist(df["age"], bins=10)

plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()

# b. PIE CHART

sex_count = df["sex"].value_counts()

plt.figure(figsize=(5,5))

plt.pie(
    sex_count.values,
    labels=["Male","Female"],
    autopct='%1.1f%%'
)

plt.title("Pie Chart of Sex")

plt.show()

# c. BOX PLOT

plt.figure(figsize=(7,5))

sns.boxplot(x="sex", y="age", data=df)

plt.title("Sex vs Age")

plt.show()

# d. SCATTER PLOT

plt.figure(figsize=(6,4))

plt.scatter(df["age"], df["chol"])

plt.title("Scatter Plot: Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")

plt.show()

# e. ADD BOXPLOT TO SCATTERPLOT

plt.figure(figsize=(7,5))

sns.boxplot(x="sex", y="age", data=df)

sns.stripplot(
    x="sex",
    y="age",
    data=df,
    color="black"
)

plt.title("Boxplot + Scatterplot")

plt.show()
```
