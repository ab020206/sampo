```python
# PS - Movie Dataset Operations

import pandas as pd

# Load Dataset
movies = pd.read_csv("movies_metadata.csv", low_memory=False)
ratings = pd.read_csv("ratings_small.csv")

# 1. Clean + Merge

movies = movies[movies["id"].astype(str).str.isnumeric()]
movies["id"] = movies["id"].astype(int)

avg = ratings.groupby("movieId")["rating"].mean().reset_index()
avg.columns = ["id", "rating"]

df = pd.merge(movies, avg, on="id")

# Keep only useful columns
df = df[[
    "id",
    "title",
    "original_language",
    "rating",
    "vote_average",
    "vote_count"
]]

print("Original Dataset")
display(df.head())

# 2. Subsets

english = df[df["original_language"] == "en"]
french = df[df["original_language"] == "fr"]

print("English Movies")
display(english.head())

# 3. Merge Subsets

merged = pd.concat([english, french])

print("Merged Data")
display(merged.head())

# 4. Sort Data

sorted_data = df.sort_values(by="rating", ascending=False)

print("Sorted Data")
display(sorted_data.head())

# 5. Transpose

print("Transpose Data")
display(df.head().T)

# 6. Melt

melted = pd.melt(
    df.head(),
    id_vars=["id", "title"],
    var_name="attribute",
    value_name="value"
)

print("Melted Data")
display(melted.head())

# 7. Pivot Table

pivot = df.pivot_table(
    values="rating",
    index="original_language",
    aggfunc="mean"
)

print("Pivot Table")
display(pivot)
```
