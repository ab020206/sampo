```python
# PS 15 - Breast Cancer Prediction

# Run:
# pip install pandas numpy scikit-learn
# python ps15.py

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score, classification_report

# Load Dataset
df = pd.read_csv("BreastCancerWc.csv", header=None)

df.columns = [
    "id","clump","cell_size","cell_shape","adhesion",
    "epithelial","bare_nuclei","chromatin",
    "nucleoli","mitoses","class"
]

print("First 5 rows:")
print(df.head())

# -----------------------------
# Data Cleaning
# -----------------------------

# Replace ? with NaN
df.replace("?", np.nan, inplace=True)

# Convert to numeric
df = df.apply(pd.to_numeric, errors="coerce")

# Remove null values
df = df.dropna()

# Remove duplicate rows
df = df.drop_duplicates()

# Remove negative values
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    df = df[df[col] >= 0]

print("\nDataset after cleaning:")
print(df.shape)

# -----------------------------
# Outlier Removal
# -----------------------------

# Not removing outliers because
# values 1-10 are valid medical scores

print("\nDataset after outlier removal:")
print(df.shape)

# -----------------------------
# Data Transformation
# -----------------------------

# 2 = Benign, 4 = Malignant
df["class"] = df["class"].map({2:0, 4:1})

X = df.drop(["id","class"], axis=1)
y = df["class"]

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# -----------------------------
# Train Test Split
# -----------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# -----------------------------
# Logistic Regression Model
# -----------------------------

lr = LogisticRegression()

lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)

lr_accuracy = accuracy_score(y_test, lr_pred)

print("\nLogistic Regression Accuracy:")
print(lr_accuracy)

# -----------------------------
# Naive Bayes Model
# -----------------------------

nb = GaussianNB()

nb.fit(X_train, y_train)

nb_pred = nb.predict(X_test)

nb_accuracy = accuracy_score(y_test, nb_pred)

print("\nNaive Bayes Accuracy:")
print(nb_accuracy)

# -----------------------------
# Comparison
# -----------------------------

print("\nAccuracy Comparison")
print("Logistic Regression :", lr_accuracy)
print("Naive Bayes         :", nb_accuracy)

print("\nClassification Report - Logistic Regression")
print(classification_report(y_test, lr_pred))

print("\nClassification Report - Naive Bayes")
print(classification_report(y_test, nb_pred))
```
