```python
# PS 18 - Heart Disease Visualization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read dataset
df = pd.read_csv("/content/Cleavland.csv", header=None)

# Add column names
df.columns = [
    "age","sex","cp","trestbps","chol","fbs",
    "restecg","thalach","exang","oldpeak",
    "slope","ca","thal","target"
]

# a. HISTOGRAM

plt.figure(figsize=(6,4))

plt.hist(df["age"], bins=10)

plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()

# b. PIE CHART

target_count = df["target"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    target_count.values,
    labels=target_count.index,
    autopct='%1.1f%%'
)

plt.title("Heart Disease Diagnosis")

plt.show()

# c. BOX PLOT

plt.figure(figsize=(7,5))

sns.boxplot(
    x="target",
    y="chol",
    data=df
)

plt.title("Cholesterol By Diagnosis")

plt.show()

# d. SCATTER PLOT

plt.figure(figsize=(7,5))

plt.scatter(
    df["age"],
    df["chol"]
)

plt.title("Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")

plt.show()

# e. BOXPLOT + SCATTER

plt.figure(figsize=(7,5))

sns.boxplot(
    x="target",
    y="oldpeak",
    data=df
)

sns.stripplot(
    x="target",
    y="oldpeak",
    data=df,
    color="black",
    alpha=0.4
)

plt.title("Oldpeak By Diagnosis")

plt.show()
```
