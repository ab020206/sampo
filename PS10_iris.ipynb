```python
# PS - Iris Dataset Operations

import pandas as pd

# Load Dataset
df = pd.read_csv("/content/Iris.csv", header=None)

# Column Names
df.columns = [
    "sepal_length",
    "sepal_width",
    "petal_length",
    "petal_width",
    "species"
]

print("Original Dataset")
display(df.head())

# ---------------------------------------------------
# 1. Create Subsets
# ---------------------------------------------------

setosa = df[df["species"] == "Iris-setosa"]
versicolor = df[df["species"] == "Iris-versicolor"]

print("Setosa Data")
display(setosa.head())

# ---------------------------------------------------
# 2. Merge Subsets
# ---------------------------------------------------

merged = pd.concat([setosa, versicolor])

print("Merged Data")
display(merged.head())

# ---------------------------------------------------
# 3. Sort Data
# ---------------------------------------------------

sorted_data = df.sort_values(by="petal_length")

print("Sorted Data")
display(sorted_data.head())

# ---------------------------------------------------
# 4. Transpose Data
# ---------------------------------------------------

print("Transpose Data")
display(df.head().T)

# ---------------------------------------------------
# 5. Melt Data
# ---------------------------------------------------

melted = pd.melt(
    df.head(),
    id_vars=["species"]
)

print("Melted Data")
display(melted.head())

# ---------------------------------------------------
# 6. Pivot Table
# ---------------------------------------------------

pivot = df.pivot_table(
    values="petal_length",
    index="species",
    aggfunc="mean"
)

print("Pivot Table")
display(pivot)
```
